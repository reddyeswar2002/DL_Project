```python
import cv2
import numpy as np
import pickle
from ultralytics import YOLO

# Helper Function
def fill_global_positions_list_and_return_img(image, frame, pos_pickle_path):
    with open(pos_pickle_path, 'rb') as f:
        posList = pickle.load(f)
    global width, height
    width, height = 107, 48 # Adjust based on car parking slot size
    vacant_count = 0
    occupied_count = 0
    intensity = 255
    target_size = (1100, 820)
    image = cv2.resize(image, target_size, interpolation=cv2.INTER_LINEAR)
    frame = cv2.resize(frame, target_size, interpolation=cv2.INTER_LINEAR)

    # Check parking slot status
    for idx, pos in enumerate(posList):
        x, y = pos
        y += 50  # Adjust Y offset if needed
        imgCrop = image[y:y + height, x:x + width]
        mask = (imgCrop == 255).astype(np.uint8)
        count = cv2.countNonZero(mask)

        # Occupied
        if count > 1500:
            occupied_count += 1
            color = (0, 255, 255)  # Yellow
        # Vacant
        else:
            vacant_count += 1
            color = (0, 255, 0)  # Green

        cv2.rectangle(frame, (x, y), (x + width, y + height), color, 2)

    # Add counts at the top of the frame
    cv2.putText(
        frame,
        f"Vacant: {vacant_count} | Occupied: {occupied_count}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (255, 255, 255),
        2,
    )

    return frame

# Main Function
def process_video(vid_path, output_path, weights_path, pos_pickle_path):
    model = YOLO(weights_path)  # Load YOLOv8 model

    # Open the video file
    cap = cv2.VideoCapture(vid_path)
    if not cap.isOpened():
        print("Error: Unable to open video.")
        return

    # Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Define VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4 video
    out = cv2.VideoWriter(output_path, fourcc, fps, (1100, 820))  # Resized to (1100, 820)

    # Process each frame
    while True:
        one_d_boxes = np.zeros((frame_height, frame_width, 1), dtype=np.uint8)
        ret, frame = cap.read()
        if not ret:
            break  # Stop if video ends

        # Run inference
        results = model(frame)

        # Extract bounding boxes
        for box in results[0].boxes.xyxy:
            x1, y1, x2, y2 = map(int, box)
            one_d_boxes[y1:y2, x1:x2] = 255  # Mark detected areas

        # Overlay parking slot statuses
        frame = fill_global_positions_list_and_return_img(one_d_boxes, frame, pos_pickle_path)

        # Write frame to output video
        out.write(frame)

    cap.release()
    out.release()
    print(f"Processed video saved to {output_path}")

# Execute the Main Code
vid_path = "/content/carPark.mp4"
weights_path = "/content/runs/detect/yolov8-custom/weights/best.pt"
output_path = "/content/output.mp4"
pos_pickle_path = "/content/CarParkPos"
process_video(vid_path, output_path, weights_path, pos_pickle_path)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# Helper Function
def detect_parking_slots(model, frame):
    """Automatically detect potential parking slots in the first frame."""
    results = model(frame)
    slot_list = []

    for box in results[0].boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)  # Extract bounding box coordinates
        slot_list.append((x1, y1, x2, y2))  # Save slot coordinates

    return slot_list

def process_parking_slots(frame, slots, model):
    """Check parking slot status and draw on frame."""
    vacant_count = 0
    occupied_count = 0
    intensity = 255

    # Create a blank mask for slots
    slot_mask = np.zeros_like(frame[:, :, 0], dtype=np.uint8)

    for (x1, y1, x2, y2) in slots:
        # Extract cropped region of the slot
        slot_mask[y1:y2, x1:x2] = intensity
        slot_img = frame[y1:y2, x1:x2]

        # Run inference to check for cars in the slot
        results = model(slot_img)
        is_occupied = len(results[0].boxes) > 0

        # Determine slot status
        if is_occupied:
            occupied_count += 1
            color = (0, 255, 255)  # Yellow for occupied
        else:
            vacant_count += 1
            color = (0, 255, 0)  # Green for vacant

        # Draw the slot on the frame
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

    # Add counts at the top of the frame
    cv2.putText(
        frame,
        f"Vacant: {vacant_count} | Occupied: {occupied_count}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (255, 255, 255),
        2,
    )

    return frame

# Main Function
def process_video(vid_path, output_path, weights_path):
    model = YOLO(weights_path)  # Load YOLOv8 model

    # Open the video file
    cap = cv2.VideoCapture(vid_path)
    if not cap.isOpened():
        print("Error: Unable to open video.")
        return

    # Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Define VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4 video
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    # Detect parking slots in the first frame
    ret, first_frame = cap.read()
    if not ret:
        print("Error: Unable to read the first frame.")
        cap.release()
        return

    slots = detect_parking_slots(model, first_frame)  # Automatically detect slots

    # Process each frame
    while True:
        ret, frame = cap.read()
        if not ret:
            break  # Stop if video ends

        # Update frame with parking slot statuses
        frame = process_parking_slots(frame, slots, model)

        # Write frame to output video
        out.write(frame)

    cap.release()
    out.release()
    print(f"Processed video saved to {output_path}")

# Execute the Main Code
vid_path = "/content/carPark.mp4"
weights_path = "/content/runs/detect/yolov8-custom/weights/best.pt"
output_path = "/content/output.mp4"
process_video(vid_path, output_path, weights_path)
```
